```python
df_indygo_dropped['ArriveBoards'] = df_indygo_dropped['ArriveBoards'].astype(int)
df_indygo_dropped['ArriveAlights'] = df_indygo_dropped['ArriveAlights'].astype(int)

df_indygo_dropped['ArriveOnboard'] = df_indygo_dropped['ArriveOnboard'].astype(int)
df_indygo_dropped['DepartBoards'] = df_indygo_dropped['DepartBoards'].astype(int)
df_indygo_dropped['DepartAlights'] = df_indygo_dropped['DepartAlights'].astype(int)
df_indygo_dropped['DepartOnboard'] = df_indygo_dropped['DepartOnboard'].astype(int)
df_indygo_dropped['DepartTimeVarianceSecs'] = df_indygo_dropped['DepartTimeVarianceSecs'].astype(int)
df_indygo_dropped['ArriveTimeVarianceSecs'] = df_indygo_dropped['ArriveTimeVarianceSecs'].astype(int)
df_indygo_dropped['EndTimeVarianceSecs'] = df_indygo_dropped['EndTimeVarianceSecs'].astype(int)
df_indygo_dropped['TotalTimeActualSecs'] = df_indygo_dropped['TotalTimeActualSecs'].astype(int)
df_indygo_dropped['TotalTimeVarianceSecs'] = df_indygo_dropped['TotalTimeVarianceSecs'].astype(int)
```
